```python
import pandas
import matplotlib.pyplot as plt
```

```python
df = pandas.read_json('bgg_data.json',orient='split')
df
```

```python
df['year'].plot.hist(bins=122,range=(1900,2022))
ax = plt.gca()
ax.set_ylabel('Number of released games / year')
ax.set_xlabel('year')
plt.savefig('ngames_1900.png')
```

```python
df['year'].plot.hist(bins=32,range=(1990,2022))
ax = plt.gca()
ax.set_ylabel('Number of released games / year')
ax.set_xlabel('year')
plt.savefig('ngames_1990.png')
```
