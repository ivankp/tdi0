```python
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
def plot_grid(points):
    points = np.array(points)
    ux = np.unique(points.T[0])
    uy = np.unique(points.T[1])
    nx = len(ux)
    ny = len(uy)
    xi = np.arange(nx)
    yi = np.arange(ny)
    
    w = np.full((nx,ny),np.nan)
    for x,y,z in points:
        w[np.searchsorted(ux,x),np.searchsorted(uy,y)] = z
    
    plt.hist2d(
        *np.array(np.meshgrid(xi,yi)).reshape(2,-1),
        weights = w.T.ravel(),
        bins = (nx,ny),
        range = ((0,nx),(0,ny))
    )

    ax = plt.gca()
    ax.set_xticks(xi+0.5)
    ax.set_yticks(yi+0.5)
    ax.set_xticklabels(f'{x:g}' for x in ux)
    ax.set_yticklabels(f'{x:g}' for x in uy)
```
